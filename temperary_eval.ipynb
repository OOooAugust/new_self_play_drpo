```python
import pandas as pd
```

```python
data = []
for temp in [0, 0.25, 0.5, 0.75, 1.0]:
    try:
        df = pd.read_csv(f"ppo_drpo_hightemp_temp_{str(temp).replace('.', '_')}_results_5.csv")
        data.append(df)
    except FileNotFoundError:
        print(f"File not found for temp {temp}")
        continue
```

```python
def calc_mean_win_rate(df):
    # remove rows where judge_result is 2 or -1
    # df = df[df['judge_result'] != 2 or df['judge_result'] != -1]
    return df['judge_result'].mean()
```

```python
[calc_mean_win_rate(df) for df in data]
```
